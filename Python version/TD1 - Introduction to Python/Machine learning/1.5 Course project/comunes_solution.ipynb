### **D. Probabilities**

**D1. Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.**

The probability that we are looking for may be computed as the reason between: the number of communes in Valais that are over 2000 m, over the total amount of communes that are over 2000 m of elevation. For extracting this information from the DF the following will be made.

```python
# Number of communes over 2000 m separated by Canton
more_than_2000 = [x > 2000 for x in df["Elevation"]]
count = df[more_than_2000].groupby("Canton").size()
count
```

```python
# Numerator = Number of communes in Valais over 2000 m in elevation
valais = count["VS"]
# Denominator = total number of communes over 2000 m  in elevation
total_communes = count.sum()
# The probability is computed as follows:
probability = valais / total_communes
print(
    "The probability of selecting a commune from the canton of Valais when randomly selecting a commune with elevation over 2000 m is: "
    + str(probability)
)
```
