(consult the documentation for the matplotlib.pyplot.scatter() method to obtain the correct syntax and list of parameters.)

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
from sklearn import datasets

iris = datasets.load_iris()
# Line #2
iris_df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
iris_df["species"] = iris["target"]

#3
iris_df.head()

#4 
species = ["Iris setosa", "Iris versicolor", "Iris virginica"]
colors  = ["red", "orange", "green"]
# Alternatively, we can use code to extract species names (instead of hard-coding them).
species = ["Iris " + x for x in iris.target_names]

```

```python
iris_df.head(5)
```

```python
df_setosa = iris_df.loc[iris_df.species==0]
df_versicolor = iris_df.loc[iris_df.species==1]
df_virginica = iris_df.loc[iris_df.species==2]
```

```python
plt.scatter(df_setosa["sepal length (cm)"], df_setosa["petal length (cm)"], color="red", edgecolor="white")
plt.scatter(df_versicolor["sepal length (cm)"], df_versicolor["petal length (cm)"], color="orange", edgecolor="white")
plt.scatter(df_virginica["sepal length (cm)"], df_virginica["petal length (cm)"], color="green", edgecolor="white")
plt.ylabel("Petal Length (cm)")
plt.xlabel("Sepal Length (cm)")
plt.title("Iris dataset: petal length vs sepal length", fontsize=12)
plt.legend(["Iris setosa", "Iris versicolor", "Iris virginica"], loc="lower right")
plt.show()
```

```python
fig, (ax1, ax2, ax3) =plt.subplots(1,3, figsize = (12, 3.5), sharex=True, sharey=True)
ax1.scatter(df_setosa["sepal length (cm)"], df_setosa["petal length (cm)"],alpha=0.7)
ax2.scatter(df_versicolor["sepal length (cm)"], df_versicolor["petal length (cm)"], alpha=0.7)
ax3.scatter(df_virginica["sepal length (cm)"], df_virginica["petal length (cm)"], alpha=0.7)
ax1.set_title('Iris setosa')
ax2.set_title('Iris versicolor')
ax3.set_title('Iris virginica')
ax1.set_xlim(4.0,8.2)
ax1.set_ylim(0.5,7.5)
fig.tight_layout()
# Add x/y labels and titles
ax1.set_xlabel('sepal length (cm)')
ax1.set_ylabel('petal length (cm)')
ax2.set_xlabel('sepal length (cm)')
ax3.set_xlabel('sepal length (cm)')
```

```python
# Alternatively and option is generating a loop for plotting, using the zip method
fig, axes =plt.subplots(1,3, figsize = (12, 3.5))
# con este loop es posible ingresar a cada plot de forma iterativa
for i, ax in zip(range(0, 3), axes.flatten()):
    #Filter data for each species
    species_df = iris_df[iris_df['species'] == i]
    # Populate subplots
    ax.scatter(species_df['sepal length (cm)'], species_df['petal length (cm)'], alpha = 0.7)
    
    # Add x/y labels and titles
    ax.set_xlabel('sepal length (cm)')
    ax.set_ylabel('petal length (cm)')
    ax.set_title(species[i])   
    
    # Apply limits to x/y axes
    ax.set_xlim([4.0,8.2])
    ax.set_ylim([0.5,7.5])

fig.tight_layout()
```

The zip function takes any number of iterables as arguments (arrays, lists, etc.) and returns tuples with the elements of each iterable. Below, we pass the function range(0,3) and the axes object to the zip function.
